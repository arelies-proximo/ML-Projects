#ANN IMPLEMENTATION

BACKPROPAGATION algorithm

Basic Implementation, 

- Dot Product,  input and Hidden Layer's Wieght of Neuron
- Adding of Bias
- Through an Activation Function
- Output of A.F. to input to Output Layer
- Dot product with weight of O/P Layers' neurons
- Add the bias
- Through an Activation Function of Output Layer
- Predicted output is obtained
- Find the error in this prediction
- Update weight of hidden layer
- Update weight of Ouput layer
- Continue interation till completion of specified epochs

```python
import numpy as np
```

```python
#2 hours sleep, 9 hours study -> 70 marks
#8 sleep, 7 stuyd -> 92 marks
# qsleep, 10 study -> 60 marks

x = np.array(([2,9], [8,7], [1,10]), dtype=float)
y = np.array(([70], [92], [60]), dtype=float)
```

```python
print(np.amax(x, axis=0))
  #max vals in each columns
```

```python
x = x/np.amax(x,axis=0)

y = y/100


print('x:\n', x)
print('\ny: ', y)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

#a sigmoid function
#to be used in
```

```python
def derivative_sigmoid(x):
    return x*(1-x)
```

Number of iteration **epoch** and the learning rate **l**

```python
epoch = 7000
lr = 0.2
```

Number of neurons in each layer

```python
ip_layer_neurons = 2
  #two x columns

hdn_layer_nrns = 3
  #3 rows in x

op_layer_nrns = 1
  #single o/p for each x pairs
```

Weight of HIDDEN layers.

Bias of HIDDEN Layer

```python
wh = np.random.uniform(size=(ip_layer_neurons, hdn_layer_nrns))
bh = np.random.uniform(size=(1,hdn_layer_nrns))
```

*With the help of numpy.random.uniform() method, we can get the **random samples from uniform distribution** and returns the random samples as numpy array by using this method*

```python
wh
```

```python
bh
```

Weight and bias of Output Layer

```python
wout = np.random.uniform(size=(hdn_layer_nrns, op_layer_nrns))
bout = np.random.uniform(size=(1, op_layer_nrns))
```

```python
wout
```

```python
bout
```

```python

for i in range(epoch):
    hinp1= np.dot(x,wh)
        #first hidden layer input is cal by a dot product
        #Dot of input and weight of hidden layer neurons
    hinp = hinp1 + bh 
        #hidden layer input
        #ADD THE BIAS

    hlayer_act = sigmoid(hinp)
        #result of activation from hidden will be input to output layer

    outinp1 = np.dot(hlayer_act, wout)
    outinp = outinp1 + bout
    output = sigmoid(outinp)

    #Back Propagation
    EO = y-output
        #Error output layer
    outgrad = derivative_sigmoid(output)
        #output gradient

    d_output = EO*outgrad

    EH = d_output.dot(wout.T)
        #error hidden layer
        # .Target

    hiddengrad = derivative_sigmoid(hlayer_act)

    d_hiddenlayer = EH*hiddengrad

    wout += hlayer_act.T.dot(d_output)*lr
    #bout += np.sum(d_output, axis=0, keepdims=True)*lr

    wh+= x.T.dot(d_hiddenlayer)*lr
    #bh += np.sum(d_hiddenlayer, axis=0, keepdims=True)*lr

    if i%100==0:
      print(f"\ti: {i}\nEO: \n{EO}, \nEH: \n{EH}\nOutput: \n{output}\n-----------------------------")

```

```python
print("Input: \n", x)
print("Actual Output: \n",y)
print("Predicted Output: \n", output)
```
